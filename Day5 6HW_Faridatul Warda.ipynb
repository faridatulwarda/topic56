```python
#Kalkulator BMI#
```

```python
weight = float(input())
```

```python
height = float(input())
```

```python
bmi = weight / (height/100)**2
```

```python
if bmi < 18.5 :
    kategori = "Underweight"
elif 18.5 <= bmi < 25 :
    kategori = "Normal Weight"
elif 25 <= bmi < 30 :
    kategori = "Overweight"
elif 30 <= bmi < 35 :
    kategori = "Obesity class 1"
elif 35 <= bmi < 40 :
    kategori = "Obesity class 2"
else :
    kategori = "Obesity class 3"

    
print("\nSkor BMI : %.2f (%s)" % (bmi, kategori))

if bmi < 18.5 :
    print("Rekomendasi: Naikkan berat badan dengan mengonsumsi makanan bergizi.")
elif 8.5 <= bmi < 25 :
    print("Berat badan sudah ideal. Jaga selalu!")
else :
    print("Rekomendasi:")
    print("- Turunkan berat badan, hindari makanan cepat saji.")
    print("- Jangan lupa rutin berolahraga.")
```
